Теоретически лемматизацию можно применить к столбцу, но в таком случае обработка идёт очень долго (больше часа)<br>
Поэтому Все данные из столбца <code>text</code> занесём в одну строку и к этой строке применим лемматизацию.

```python
all_text = df['text'].apply(lambda x: x + ' ').sum()
all_text
```

Лемматизацию проводим с помощью MyStem (т.к. он подходит для работы с русскими словами)

```python
from pymystem3 import Mystem
```

```python
%%time
all_text = ''.join(Mystem().lemmatize(all_text))
all_text
```

Уберём лишние знаки препинания

```python
all_text = all_text.replace('.', '').replace(',', '').replace(':', '').replace(';', '').replace('"', '').replace('(', '').replace(')', '').replace('- ', '-').replace('-и ', '- и ')
```

Теперь все слова из нашей строки занесём в список (разделитель - пробел)

```python
list_text = all_text.split(sep=' ')
list_text
```

```python
len(list_text)
```

```python
# Количество уникальных значений
len(set(list_text))
```

```python
set(list_text)
```

Создаём функцию, которая проверяет, содержит ли элемент столбца <code>text</code> элемент нашего списка <code>unique_list_stem</code>. Если да, то возвращает значение "компьютерный класс".

```python
def fun(text):
    for element in unique_list_stem:
        if element in text:
            return 'компьютерный класс'
```

Применяем функцию к столбцу <code>text</code>, записывая результат в новый столбец <code>class_type</code>

```python
df['class_type'] = df['text'].apply(fun)
```

```python
df['class_type'] = df['class_type'].fillna('обычная аудитория')
```
